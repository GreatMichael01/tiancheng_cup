```python
#  -*- coding: utf-8 -*-
import pandas as pd
import xgboost as xgb
```

### Read Data Train 1-5

```python
original_train_1 = pd.read_table('data/open_data_train_valid3/train/train_1.txt')
original_train_2 = pd.read_table('data/open_data_train_valid3/train/train_2.txt',names = list(original_train_1.columns.values))
original_train_3 = pd.read_table('data/open_data_train_valid3/train/train_3.txt',names = list(original_train_1.columns.values))
original_train_4 = pd.read_table('data/open_data_train_valid3/train/train_4.txt',names = list(original_train_1.columns.values))
original_train_5 = pd.read_table('data/open_data_train_valid3/train/train_5.txt',names = list(original_train_1.columns.values))
```

### construct aggregation of train and test

```python
original_train_aggregate = pd.concat([original_train_1,original_train_2,original_train_3,original_train_4,original_train_5])
```

```python
del original_train_1
del original_train_2
del original_train_3
del original_train_4
del original_train_5
```

```python
train_column_list = list(original_train_aggregate.columns.values)
train_column_list.remove('label')
train_column_list.remove('tag')
```

```python
original_test = pd.read_table('data/open_data_train_valid3/valid.txt')[train_column_list]
```

### Weekday Feature Creation

```python
original_train_aggregate['weekday']=pd.to_datetime(original_train_aggregate.loan_dt,format='%Y-%m-%d').dt.dayofweek
```

```python
original_test['weekday']=pd.to_datetime(original_test.loan_dt,format='%Y-%m-%d').dt.dayofweek
```

### Transfer TAG['fq','pd'] into [1,0]

```python
original_train_aggregate['tag'] = original_train_aggregate['tag'].map({'fq':int(1), 'pd': int(0)})
```

```python
original_train_aggregate
```

```python
original_train_aggregate['trans_loan_dt']=pd.to_datetime(original_train_aggregate.loan_dt,format='%Y-%m-%d')
```

```python
original_train_aggregate['YearMonth'] = original_train_aggregate['trans_loan_dt'].map(lambda x: 100*x.year + x.month)
```

```python
original_train_aggregate[(original_train_aggregate['loan_dt'] > '2018-01-01')&(original_train_aggregate['tag'] == 'fq')]
```

```python
original_train_aggregate.groupby(['YearMonth','tag','label']).id.count()
```

### original_train_x, original_train_y, original_test_x construct

```python
Jan_to_May_Tag_train_x = original_train_aggregate.drop(columns=['id','loan_dt','label','tag']).fillna(-99999)
```

```python
Jan_to_May_Tag_train_y = pd.DataFrame(original_train_aggregate['tag'].fillna(-99999))
```

```python
original_test_x = original_test.drop(columns=['id','loan_dt']).fillna(-99999)
```

### Feature Importance & Selection

```python
Jan_to_May_Tag_xgb_train = xgb.DMatrix(Jan_to_May_Tag_train_x,label = Jan_to_May_Tag_train_y)
```

```python
del Jan_to_May_Tag_train_x
```

```python
Jan_to_May_Tag_xgb_train_x = xgb.DMatrix(Jan_to_May_Tag_train_x)
```

```python
Jan_to_May_Tag_xgb_test_x = xgb.DMatrix(original_test_x)
```

```python
Jan_to_May_Tag_train_y_params = {"booster": "gbtree",
          "objective" : "binary:logistic",
          "eval_metric" : "auc",
          "verbose" : 1, 
          "eta" : 0.01, 
          "max_delta_step" : 5, 
          "max_depth" : 20, 
          "alpha" : 10, 
          "lambda" : 10, 
          "gamma" : 2, 
          "subsample" : 0.8, 
          "colsample_bytree" : 0.8, 
          "min_child_weight" : 5, 
          "scale_pos_weight" : 0.8}
Jan_to_May_Tag_num_rounds = 3000
```

```python
Jan_to_May_Tag_xgb_model = xgb.train(Jan_to_May_Tag_train_y_params,Jan_to_May_Tag_xgb_train,Jan_to_May_Tag_num_rounds)
```

```python
train_x_tag = Jan_to_May_Tag_xgb_model.predict(Jan_to_May_Tag_xgb_train_x)
```

```python
test_x_tag =  Jan_to_May_Tag_xgb_model.predict(Jan_to_May_Tag_xgb_test_x)
```

```python
Jan_to_May_Tag_train_x['pred_tag'] = pd.DataFrame(train_x_tag)
```

```python
original_test_x['pred_tag'] = pd.DataFrame(test_x_tag)
```

```python
final_train = xgb.DMatrix(Jan_to_May_Tag_train_x,label = Jan_to_May_Tag_train_y)
```

```python
final_test = xgb.DMatrix(original_test_x)
```

```python
final_model = xgb.train(Jan_to_May_Tag_train_y_params,final_train,Jan_to_May_Tag_num_rounds)
```

```python
final_prob_result = final_model.predict(final_test)
```

```python
final_result = pd.DataFrame(original_test['id'])
```

```python
final_result
```

```python
final_result['prob'] = pd.DataFrame(data = final_prob_result, columns = ['prob'])
```

```python
final_result.to_csv('data/open_data_train_valid3/result.txt',index = None, sep=',')
```
